```python
import requests
import time
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

iss_data = []

api_url = 'https://api.wheretheiss.at/v1/satellites/25544'

start_time = time.time()

# Run program for 17 minutes
while time.time() - start_time < 17 * 60:
    response = requests.get(api_url)

    if response.status_code == 200:
        data = response.json()
        latitude = data["latitude"]
        longitude = data["longitude"]
        timestamp = data["timestamp"]

    # Append data
        iss_data.append((timestamp, latitude, longitude))

        time.sleep(10)  # 10-second intervals
    else:
        print("error")
        break
        
# Plotting the points
timestamps, latitudes, longitudes = zip(*iss_data)
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(longitudes, latitudes, color='blue')

# Set labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('ISS path')
ax.grid(True)

# Show the plot
plt.show()
```
